### McCulloch-Pitts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def activation_function(x):
    return 1 if x >= 1 else 0
```

```python
def McCulloch_Pitts(x, y, t, w, b):
    """
    McCulloch-Pitts neuron model.
    
    Parameters:
    x : list
        Input values.
    y : list
        Output values.
    t : list
        Target values.
    w : list
        Weights for the inputs.
    b : int 
        Bias.
    """
    t_preds = []
    for i in range(len(x)):
        yin = w[0] * x[i] + w[1] * y[i] + b
        t_pred = activation_function(yin)
        t_preds.append(t_pred)
        print(f"Input: ({x[i]}, {y[i]}), Weighted sum: {yin}, Predicted: {t_pred}, Target: {t[i]}")

    return t_preds  
```

```python
def plotMcCulloch_Pitts(x, y, t, w, b):
   plt.scatter(x, y, c=t, cmap='coolwarm', marker='s')
   # Equation of desction boundary
   X = np.linspace(-0.2, 1.2, 100)
   Y = [-xi*(w[0]/w[1]) - b/w[1] for xi in X]
   plt.plot(X, Y, color='black', label=f'y={-w[0]/w[1]}x - {b/w[1]}')
   plt.legend()
   plt.xlabel('x')
   plt.ylabel('y')
   plt.title('McCulloch-Pitts Decision Boundary for b = {} w = [{},{}]'.format(b, w[0], w[1]))
   plt.grid(True, alpha=0.3)
   plt.figure(figsize=(4, 4)) 
```

#### AND Gate

```python
x = [0, 0, 1, 1]
y = [0, 1, 0, 1]
t = [0, 0, 0, 1]
```

```python
w = [1, 1]  # Weights for x and y
b = 1  # Bias
t_preds = McCulloch_Pitts(x, y, t, w, b)
```

```python
w = [1, 1]  # Weights for x and y
b = -1  # Bias
t_preds = McCulloch_Pitts(x, y, t, w, b)
```

```python
w = [1, 1]  # Weights for x and y
bias_values = [1, -1, -1.5, -2]

for b in  bias_values:
    plotMcCulloch_Pitts(x, y, t, w, b)

```

#### OR Gate

```python
x = [0, 0, 1, 1]
y = [0, 1, 0, 1]
t = [0, 1, 1, 1]
```

```python
b = 0
t_preds = McCulloch_Pitts(x, y, t, w, b)
```

```python
w = [1, 1]  # Weights for x and y
bias_values = [0, -1, -1.5, -2]
for b in  bias_values:
    plotMcCulloch_Pitts(x, y, t, w, b)
```
