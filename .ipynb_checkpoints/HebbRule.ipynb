```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def activationFunction(x):
    return 1 if x >= 0 else -1
```

```python
def plotDescisionBoundary(x, y, t, w, b, idx):
   plt.subplot(8,8,idx)
   plt.scatter(x, y, c=t, cmap='coolwarm', marker='s')
   # Equation of desction boundary
   X = np.linspace(-1.2, 1.2, 100)
   Y = [-xi*(w[0]/w[1]) - b/w[1] for xi in X]
   plt.plot(X, Y, color='black', label=f'y={-w[0]/w[1]}x - {b/w[1]}')
   plt.title(f'Iteration {idx}')
```

```python
def HebbianLearning(x,y,t,w,b):
    """
    Hebbian learning rule.
    
    Parameters:
    x : numpy array
        Input vector.
    y : numpy array
        Output vector.
    t : numpy array
        target vector.
    w : numpy array
        Weight matrix.
    b : float
        Bias term.

    Returns:
    numpy array
        Updated weight matrix after applying Hebbian learning.
    """

    plt.figure(figsize=(16, 16))
    
    for i in range(len(x)):
        yin = w[0]*x[i] + w[1]*y[i] + b
        t_pred = activationFunction(yin)

        w1new = w[0] + x[i]*t[i]
        w2new = w[1] + y[i]*t[i]
        bnew = b + t[i]

        w = [w1new, w2new]
        b = bnew
        
        print(f'Updated weights: {w}, {b} t = {t[i]} t_pred = {t_pred}')
        plotDescisionBoundary(x, y, t, w, b, i+1)
    
    return w, b
```

### AND Gate

```python
w = [1 , 1]
b = 1
x = np.array([0, 0, 1, 1])
y = np.array([0, 1, 0, 1])
t = np.array([0, 0, 0, 1])  # Target
w, b = HebbianLearning(x, y, t, w, b)
print(f'Final Weights: {w}, Bias: {b}')
```

### POLAR AND GATE

```python
w = [1 , 1] 
b = 1
x = np.array([-1, -1, 1, 1])
y = np.array([-1, 1, -1, 1])
t = np.array([-1, -1, -1, 1])  # Target
w, b = HebbianLearning(x, y, t, w, b)
print(f'Final Weights: {w}, Bias: {b}')
```

### Polar OR Gate

```python
w = [1, 1] 
b = 1
x = np.array([-1, -1, 1, 1])
y = np.array([-1, 1, -1, 1])
t = np.array([-1, 1, 1, 1]) 
w, b = HebbianLearning(x, y, t, w, b)
print(f'Final Weights: {w}, Bias: {b}')
```
