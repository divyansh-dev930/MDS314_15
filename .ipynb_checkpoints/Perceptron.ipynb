```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def activationFunction(x):
    return 1 if x >= 0 else -1
```

```python
def plotDescisionBoundary(x, y, t, w, b, idx):
   plt.subplot(4,4,idx)
   plt.scatter(x, y, c=t, cmap='coolwarm', marker='s')
   # Equation of desction boundary
   X = np.linspace(-max(x), max(x), 100)
   if(w[1]==0):
      plt.title('No Boundary Possible')
      return
   Y = [-xi*(w[0]/w[1]) - b/w[1] for xi in X]
   plt.plot(X, Y, color='black', label=f'y={-w[0]/w[1]}x - {b/w[1]}')
   plt.title(f'Epoch {idx}')
```

```python
def Perceptron(x, y, t, w, b, alpha, max_iter):
    """
    Perceptron learning algorithm.
    
    Parameters:
    x : list of lists
        Input features.
    y : list
        Target labels.
    t : int
        Number of features.
    w : list
        Initial weights.
    b : float
        Initial bias.
    alpha : float
        Learning rate.
    max_iter : int
        Maximum number of iterations.
    
    Returns:
    w : list
        Updated weights after training.
    b : float
        Updated bias after training.
    """
    plt.figure(figsize=(12,12))
    for _ in range(max_iter):
        print(f"Epoch {_+1}")
        error = 0
        for i in range(len(x)):
            yin = x[i]*w[0] + y[i]*w[1] + b
            t_pred = activationFunction(yin)
            print(f"Target: {t[i]}, Predicted: {t_pred}, Input: ({w[0]}, {w[1]}) Bias: {b}")
            if(t_pred !=t[i]):
                w[0] = w[0] + alpha* x[i]*t[i]
                w[1] = w[1] + alpha* y[i]*t[i]
                b = b + alpha * t[i]
                error += 1
                
        plotDescisionBoundary(x, y, t, w, b, _ + 1)
        if error == 0:
            print("No errors, stopping training.")
            break
        else:
            print(f"Errors in this epoch: {error}")

    plt.plot()
    plt.show()
        
    
    return w, b
```

### Polar AND gate

```python
x = [1, 1, -1, -1]
y = [1, -1, 1, -1]
t = [1, -1, -1, -1]  # Polar AND gate
w = [0, 0]
b = 0
alpha = 1
max_iter = 5
w, b = Perceptron(x, y, t, w, b, alpha, max_iter)
```

### Polar OR Gate

```python
x = [1, 1, -1, -1]
y = [1, -1, 1, -1]
t = [1, 1, 1, -1] 
w = [0, 0]
b = 0  
alpha = 1
max_iter = 12
w, b = Perceptron(x, y, t, w, b, alpha, max_iter)
```
